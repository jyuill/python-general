## Filtering

```python
# specify conditions for one or more cols AND
df[(df["name"]== "External Withdrawal Mortgage") & (df["amount"] < 0)]
```

```python
# multiple filter values same column OR
df[(df["name"]== "External Withdrawal Mortgage") | (df["name"]== "External Deposit Rent/Leases")]
```

```python
# filter based on multiple columns OR
# - not clear how the ( ) brackets work - first ones needed, not sure about second
df[(df["name"]== "External Withdrawal Mortgage") | (df["memo"].str.contains("VANTAGE WEST"))]
```

```python
# filter on string CONTAINS
df[df["name"].str.contains("External")][0:10] # get first 10
```

```python
# filter on string CONTAINS 
# account for missing values
df[(df["memo"].str.contains("VANTAGE WEST", na=False))]
```

```python
# filter on string STARTS WITH
# - show only first 5
df[df["name"].str.startswith("Point")][0:5]
```

```python
# with values in a list
df[df["category"].isin(["rental inc","supplies"]) & (df["date"] > "2025-07-01T00:00:00Z")]
```

```python
# NOT match
df[~(df["type"] == "DEBIT")][0:4]
```

## UPDATE

```python
# Find with loc - similar to filter but allows operations
df.loc[df["name"] == "Kindle Svcs"]
```

```python
# change category
df.loc[df["name"] == "Kindle Svcs", "category"] = "books"
```

```python
# check
df[df["name"] == "Kindle Svcs"]
```

```python
# replace string
df["category"] = df["category"].str.replace("rental inc", "rental")
```

```python
# format: upper case
df["category"] = df["category"].str.upper()
# check
df[0:4]
```

## Cleaning Data

```python
# check for missing values (NaN)
df.isna().sum()
```

```python
# remove rows with missing values
#df.dropna(inplace=True) # will drop ALL rows, since at least one NaN in each row
# NA dropped for specific col
df.dropna(subset=["memo"], inplace=True) # inplace=TRUE to modify existing df
# check
df.isna().sum()
```

```python
# fill missing values
#df["memo"].fillna("no memo", inplace=True) # old method being deprecated
df.fillna({"memo": "no memo"}, inplace=True)
# check
df.isna().sum()
```

```python
# check for NA replacement
df[df["memo"] == "no memo"]
```
